# Encoding sex column

```python
Titanic['Sex']=Titanic['Sex'].replace('male','M')
Titanic['Sex']=Titanic['Sex'].replace('female','F')
Titanic.head()
```

```python
Titanic['Sex'].value_counts()
```

```python
tit=df.drop(['Sex'],axis='columns')
tit
```

# Family size

we will create a new column by the name of family which will be the sum of SibSp and parch cols

```python
tit['family_type']=tit['SibSp']+tit['Parch']
```

```python
tit.head(3)
```

```python
def Family_Size(number):
    if number==0:
        return 'Alone'
    elif number>0 and number<=4:
        return 'Medium'
    else:
        return 'Large'
```

```python
tit['Family_Size']=tit['family_type'].apply(Family_Size)
tit.head(3)
```

```python
df2=tit.drop(['family_type','SibSp','Parch'],axis='columns')
df2
```

```python
df2.dtypes
```

since room has missing NaN values it cannot be converted from float to int. if we do astype() it will show ValueError: cannot convert float(Nan or inf) to int
that's the reason we are using fillna()

```python
df2['Room']=df2['Room'].fillna(df2['Room'].mode()[0])
df2
```

converting Room data type to int 

```python
df2['Room']=df2['Room'].astype(int)
df2
```

# One Hot Encoding on all categorical variables 

```python
titanic=pd.get_dummies(df2,columns=['Name','Embarked','Title','Deck','Family_Size','Ticket'])
titanic
```
